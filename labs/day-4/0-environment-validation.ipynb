# Environment validation

Run this notebook first from the `labs/day-4` folder. It checks Python 3.11, installs dependencies from `requirements.txt`, and confirms your OpenAI key works.

On **Windows**, `requirements.txt` also installs `pywin32`, which the MCP client needs for stdio subprocess handling.

```python
import sys, subprocess

subprocess.run(
    [sys.executable, "-m", "pip", "install", "-q", "-r", "requirements.txt"],
    stderr=subprocess.DEVNULL,
    check=False,
)
if sys.platform == "win32":
    subprocess.run(
        [sys.executable, "-m", "pip", "install", "-q", "pywin32>=306"],
        stderr=subprocess.DEVNULL,
        check=False,
    )
print("Dependencies OK (see requirements.txt).")
```

```python
import sys
print("Python:", sys.version)
major, minor = sys.version_info[:2]
assert major == 3 and minor == 11, f"Need Python 3.11.x (found {major}.{minor})"
print("Python OK")
```
